````python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import kagglehub
from sklearn.model_selection import train_test_split
````

````python
import joblib
````

````python
import os

path = kagglehub.dataset_download("neurocipher/student-performance")
files = os.listdir(path)
df = pd.read_csv(os.path.join(path, files[0]))
````

````python
nrows = df.shape[1]
ncol = df.shape[0]
print(nrows,
ncol)
````

````python
df.head()
````

````python
mapping = {"Yes":1, "No":0}
df["Extracurricular Activities"] = df["Extracurricular Activities"].map(mapping)
df.head()
````

````python
x_columns = ["Hours Studied", "Previous Scores", "Extracurricular Activities", "Sleep Hours", "Sample Question Papers Practiced"]
````

````python
for x_column in x_columns:
  df.plot(kind = "scatter", x = x_column, y = "Performance Index")
plt.show()
````

````python
x = df[x_columns]
y = df["Performance Index"]
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.1, random_state = 30)
````

````python
model = LinearRegression()
model.fit(x_train, y_train)
````

````python
rsq = model.score(x_test, y_test)
print(f"model score : {rsq}")
````

````python
x_columns_1 = x_columns[:3]
x_1 = df[x_columns_1]
y_1 = df["Performance Index"]
x_1_train, x_1_test, y_1_train, y_1_test = train_test_split(x_1, y_1, test_size = 0.1, random_state = 30)
model_1 = LinearRegression()
model_1.fit(x_1_train, y_1_train)

rsq_1 = model_1.score(x_1_test, y_1_test)
print(f"model score : {rsq_1}")
````

````python
x_columns_3 = x_columns[:3] + x_columns[4:]
x_3 = df[x_columns_3]
y_3 = df["Performance Index"]
x_3_train, x_3_test, y_3_train, y_3_test = train_test_split(x_3, y_3, test_size = 0.1, random_state = 30)
model_3 = LinearRegression()
model_3.fit(x_3_train, y_3_train)

rsq_3 = model_3.score(x_3_test, y_3_test)
print(f"model score : {rsq_3}")
````

````python
x_columns_4 = x_columns[2:]
x_4 = df[x_columns_2]
y_4 = df["Performance Index"]
x_4_train, x_4_test, y_4_train, y_4_test = train_test_split(x_4, y_4, test_size = 0.1, random_state = 30)
model_4 = LinearRegression()
model_4.fit(x_4_train, y_4_train)

rsq_4 = model_4.score(x_4_test, y_4_test)
print(f"model score : {rsq_4}")
````
